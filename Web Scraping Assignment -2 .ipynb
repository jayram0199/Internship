have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10

2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
driver =webdriver.Chrome(r"chromedriver.exe")
```

```python
# Opening the Naukari page
driver.get("https://www.naukri.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys("Data Analyst")
```

```python
location =driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bangalore")
```

```python
search = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title =[]
job_location =[]
company_name= []
exprience_required =[]
```

```python
title_tags = driver.find_elements(By.XPATH,"//a[@class='title ellipsis']")
for i in title_tags[0:10]:
    title = i.text
    job_title.append(title)
location_tags = driver.find_elements(By.XPATH,"//span[@class='ellipsis fleft locWdth'] ")
for i in location_tags[0:10]:
    location =i.text
    job_location.append(location)
company = driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft'] ")
for i in company [0:10]:
    com = i.text
    company_name.append(com)
exprience =driver.find_elements(By.XPATH,"//span[@class='ellipsis fleft expwdth'] ")
for i in exprience[0:10]:
    exe = i.text
    exprience_required.append(exe)

```

```python
Data = pd.DataFrame({"job_title":job_title,"job_location":job_location,"company_name":company_name,"exprience_required":exprience_required})
Data
```

```python
Brand = []
ProductDescription = []
Price = []
```

```python
for page in range(3):
    titles = driver.find_elements(By.XPATH,"//div[@class='_2WkVRV']")
    for i in titles:
         Brand.append(i.text)
    PD = driver.find_elements(By.XPATH,"//a[@class='IRpwTa']")
    for i in PD :
        ProductDescription.append(i.text)
    price =driver.find_elements(By.XPATH,"//div[@class ='_30jeq3']")
    for i in price:
        Price.append(i.text)
    driver.execute_script("scrollBy(0,500)")    
    next_button = driver.find_element(By.XPATH,"//*[text() ='Next']")
    next_button.click()
    time.sleep(3)
    
```

```python
Data=pd.DataFrame()
Data["BRAND"]=Brand[:100]
Data["PRODUCT_DESCRIPTION"]=ProductDescription[:100]
Data["PRICE"]=Price[:100]
Data
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
Top_Quotes = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
Top_Quotes.click()
```

```python
Quote =[]
Auther =[]
Type_Of_Quote =[]
```

```python
for page in range(10):
    titles = driver.find_elements(By.XPATH,"//a[@class='title']")
    for i in titles:
         Quote.append(i.text)
    PD = driver.find_elements(By.XPATH,"//div[@class='author']")
    for i in PD :
        Auther.append(i.text)
    price =driver.find_elements(By.XPATH,"//div[@class='tags']")
    for i in price:
        Type_Of_Quote.append(i.text)
    next_button = driver.find_element(By.XPATH, "/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]")
    next_button.click()
    time.sleep(3)
```
